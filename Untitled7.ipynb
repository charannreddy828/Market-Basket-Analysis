```python
# Import necessary libraries for data manipulation, visualization, and frequent pattern mining
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
# Load the dataset from the specified CSV file into a pandas DataFrame
df = pd.read_csv('store_data_git.csv', header=None)
# Display the first few rows of the DataFrame to inspect the data
df.head()
```

```python
# Convert the pandas DataFrame into a list of lists, where each inner list represents a single transaction
basket = []
for index, row in df.iterrows():
    basket.append(row.dropna().tolist())

# Use the TransactionEncoder to transform the list of transactions into a one-hot encoded DataFrame
# This format is required for the Apriori algorithm
te = TransactionEncoder()
te_ary = te.fit(basket).transform(basket)
df_encoded = pd.DataFrame(te_ary, columns=te.columns_)
# Display the first few rows of the one-hot encoded DataFrame
df_encoded.head()
```

```python
# Apply the Apriori algorithm to find frequent itemsets in the one-hot encoded DataFrame
# min_support=0.01 means that an itemset must appear in at least 1% of the transactions to be considered frequent
# use_colnames=True allows the output DataFrame to use the actual item names as column headers
frequent_itemsets = apriori(df_encoded, min_support=0.01, use_colnames=True)
# Display the resulting frequent itemsets
frequent_itemsets
```

```python
# Sort the frequent itemsets by their support in descending order and select the top 20
frequent_itemsets_top20 = frequent_itemsets.sort_values(by='support', ascending=False).head(20)
# Reset the index of the resulting DataFrame
frequent_itemsets_top20 = frequent_itemsets_top20.reset_index(drop=True)
# Display the top 20 frequent itemsets
frequent_itemsets_top20
```

```python
# Filter the frequent itemsets to include only those with a length of 1 (single items)
top_pairs_1 = frequent_itemsets[frequent_itemsets['itemsets'].apply(lambda x: len(x) == 1)].sort_values(by='support', ascending=False)
# Display the single-item frequent itemsets sorted by support
top_pairs_1
```

```python
# Get the top 10 1-item frequent itemsets sorted by support for visualization
top_pairs = frequent_itemsets[frequent_itemsets['itemsets'].apply(lambda x: len(x) == 1)].sort_values(by="support", ascending=False).head(10)

# Convert the itemsets (frozensets) to a comma-separated string for plotting
top_pairs['itemsets'] = top_pairs['itemsets'].apply(lambda x: ', ' .join(list(x)))
pairs = top_pairs['itemsets']
# Calculate the support values as percentages for plotting
support_values = top_pairs['support']*100

# Create a horizontal bar chart to visualize the top 10 1-item frequent pairs
plt.figure(figsize=(10, 6))
bars = plt.barh(pairs, support_values, color='lightblue')
# Add text labels to the bars showing the support percentage
for bar in bars:
  plt.text(bar.get_width() + 0.1, bar.get_y() + bar.get_height()/2,
     f'{bar.get_width(): .2f}%', va='center', ha='left', color='black', fontsize=10)

# Set the labels and title of the plot
plt.xlabel('Support (% of Orders)')
plt.ylabel('Pair')
plt.title('Top 10 1-item Frequent Pairs')
# Remove the top and right spines for a cleaner look
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
# Invert the y-axis to display the highest support at the top
plt.gca().invert_yaxis()
# Display the plot
plt.show()
```

```python
# Generate association rules from the frequent itemsets using 'lift' as the metric and a minimum threshold of 1
raw_rules = association_rules(frequent_itemsets, metric='lift', min_threshold=1)
# Select and reorder relevant columns for the association rules, then sort by lift in descending order
rules = raw_rules[['antecedents', 'consequents', 'antecedent support', 'consequent support', 'support', 'confidence', 'lift']].sort_values(by='lift', ascending=False)
# Display the generated association rules
rules
```

```python
# Create a new column 'rule' that represents the association rule in a readable format (antecedents --> consequents)
rules['rule'] = rules['antecedents'].apply(lambda x: ', '.join(sorted(x))) + " --> " + rules ['consequents'].apply(lambda x: ', '.join(sorted(x)))

# Reorder the columns to place the 'rule' column at the beginning
rules = rules[['rule'] + [col for col in rules.columns if col != 'rule']]

# Reset the index of the DataFrame
rules = rules.reset_index(drop=True)
# Display the association rules with the new 'rule' column
rules
```

```python
# Filter the association rules to show only those with a lift greater than 1 and a confidence greater than 0.2
# Lift > 1 indicates that the antecedent and consequent are more likely to occur together than separately
# Confidence > 0.2 means that at least 20% of transactions containing the antecedent also contain the consequent
rules[(rules['lift'] > 1) & (rules['confidence'] > 0.2)]
```

```python
# Select the 'rule' and 'lift' values for the top 10 rules based on lift for visualization
rule = rules['rule'].head(10)
lift_values = rules['lift'].head(10)

# Create a horizontal bar chart to visualize the top 10 rules by lift
plt.figure(figsize=(8, 6))
bars = plt.barh(rule, lift_values, color='lightblue')
# Add text labels to the bars showing the lift value
for bar in bars:
  plt.text(bar.get_width() + 0.1, bar.get_y() + bar.get_height()/2,.
    f'{bar.get_width():.2f}', va='center', ha='left', color='black', fontsize=10)

# Set the labels and title of the plot
plt.xlabel('Lift')
plt.ylabel('Rule')
plt.title('Top 10 Rules by Lift')
# Remove the top and right spines for a cleaner look
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
# Invert the y-axis to display the highest lift at the top
plt.gca().invert_yaxis()
# Display the plot
plt.show()
```
